```python
corr_mat=df_numerized.corr()
corr_mat
```

```python
# unstacking the matrix
corr_pair=corr_mat.unstack()
corr_pair.head(45)
```

```python
#sorting the matrix

sorted_pairs=corr_pair.sort_values()
sorted_pairs
```

```python
#high Correlation
high_corr=sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
